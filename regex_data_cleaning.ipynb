```python
#define a function to loop through a column and find if there is a match with the regex pattern

def re_clean(pattern, df, pattern_name):
  pattern_en = []
  pattern_rw = []
  for row in range(len(df)): #row contains the sentence
    found_en = re.findall(pattern, df.loc[row]['english']) #scan the english column
    found_rw = re.findall(pattern, df.loc[row]['kinyarwanda']) #scan the 
    #if a match is found then process to remove it
    if found_en:
      pattern_en.append(True)  #English row contains the pattern
    else:
      pattern_en.append(False)
    if found_rw:
      pattern_rw.append(True)
    else:
      pattern_rw.append(False)
  return pattern_en, pattern_rw
```

```python
# ** A-Z, a-z, /d, .- ** @ ** /w  ** . ** /w

email_re = "\S:\S+\.\S+\.\S+|[\S]+@\S+\.\S+|[\S]+@\S+\.\s\S+\.\s\S+"
url_re = "http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
special_re = "[@_!#$%^&*()<>?/\|}{~:;-]" #List unicode characters 
```

```python
df['email_en'], df['email_rw'] = re_clean(email_re, df, "email")
```

```python
def re_sub(pattern, df=df):
  new_rw = []
  new_en = []
  for row in range(len(df)):
    new_rw.append(re.sub(pattern, '', df['kinyarwanda'][row]))
    new_en.append(re.sub(pattern, '', df['english'][row]))

  return new_en, new_rw
```

```python
df['english'], df['kinyarwanda'] = re_sub(email_re)
df['english'], df['kinyarwanda'] = re_sub(url_re)
df['english'], df['kinyarwanda'] = re_sub(special_re)
df['english'], df['kinyarwanda'] = re_sub('[\.]')
```

```python
df['email_en'], df['email_rw'] = re_clean(email_re, df, "email")
df['url_en'], df['url_rw'] = re_clean(url_re, df, "email")
df['special_en'], df['special_rw'] = re_clean(special_re, df, "email")
df['punc_en'], df['punc_rw'] = re_clean('[\.]', df, "email")
```
